```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("ggplot")
```

```python
df = pd.read_csv("metrics_whole40_whole.csv", sep=";")
whole40_train = df.query("set == 'train'")[0:-1:2]
whole40_test = df.query("set == 'val'")


df = pd.read_csv("metrics_decoder40_decoder.csv", sep=";")
decoder40_train = df.query("set == 'train'")[0:-1:2]
decoder40_test = df.query("set == 'val'")

df = pd.read_csv("metrics_none40_none.csv", sep=";")
none40_train = df.query("set == 'train'")[0:-1:2]
none40_test = df.query("set == 'val'")
```

```python
fig, ax = plt.subplots(1,2, figsize=(15,6))
plt.style.use('ggplot')
plt.style.use('default')
plt.rcParams['axes.facecolor'] = 'white'

ax[0].plot(whole40_train["loss"], label="whole model")
ax[0].plot(decoder40_train["loss"], label="decoder model")
ax[0].plot(none40_train["loss"], label="not trained")
ax[0].set_title("train/loss", fontsize=14, y=1.1)
ax[0].legend(fontsize=10, loc='upper center', bbox_to_anchor=(0.5, 1.09), ncol=3)
ax[0].grid(axis='y')

ax[1].plot(whole40_train["miou"], label="whole model")
ax[1].plot(decoder40_train["miou"], label="decoder model")
ax[1].plot(none40_train["miou"], label="not trained")
ax[1].set_title("train/mIoU", fontsize=14, y=1.1)
ax[1].legend(fontsize=10, loc='upper center', bbox_to_anchor=(0.5, 1.09), ncol=3)
ax[1].grid(axis='y')

fig.subplots_adjust(wspace=0.3, hspace=0.3)
fig.savefig("img/segformer_train.png")
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize=(15,6))
plt.style.use('ggplot')
plt.style.use('default')
plt.rcParams['axes.facecolor'] = 'white'

ax[0].plot(whole40_test["loss"], label="whole model")
ax[0].plot(decoder40_test["loss"], label="decoder model")
ax[0].plot(none40_test["loss"], label="not trained")
ax[0].set_title("val/loss", fontsize=14, y=1.1)
ax[0].legend(fontsize=10, loc='upper center', bbox_to_anchor=(0.5, 1.09), ncol=3)
ax[0].grid(axis='y')
ax[1].plot(whole40_test["miou"], label="whole model")
ax[1].plot(decoder40_test["miou"], label="decoder model")
ax[1].plot(none40_test["miou"], label="not trained")
ax[1].set_title("val/mIoU", fontsize=14, y=1.1)
ax[1].set_ylim([0,1])
ax[1].legend(fontsize=10, loc='upper center', bbox_to_anchor=(0.5, 1.09), ncol=3)
ax[1].grid(axis='y')
fig.subplots_adjust(wspace=0.3, hspace=0.3)
fig.savefig("img/segformer_test.png")
plt.show()
```
